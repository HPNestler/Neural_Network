```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdFMCS as MCS
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
df = pd.read_csv('../SF3B1_partner_mol.csv')
#print(df)
mol_df = df['smiles']
mol_df.drop_duplicates(inplace=True)
#print(mol_df)
mol_smi = mol_df.to_list()
print(mol_smi)
```

```python
mols = []
for i in mol_smi:
    #print(i)
    b = Chem.MolFromSmiles(i)
    mols.append(b)
    
print(mol_smi[0])
print(mol_smi[10])
print(mol_smi[20])

mt = [mols[0],mols[1],mols[2]]
#MCS.FindMCS(mt)
mol_mcs = MCS.FindMCS(mt)
#print(MCSResult)
print(mol_mcs.numAtoms)
print(mol_mcs.numBonds)
print(mol_mcs.smartsString)
mcs_smart = mol_mcs.smartsString

#print(dir(mol_mcs))

    
```

```python
#mols = [  m for m in sdf ]
core = Chem.MolFromSmarts(mcs_smart)
core_smi = Chem.MolToSmiles(core)
print(core_smi)
img = Draw.MolsToGridImage(mt, molsPerRow=3, highlightAtomLists=[ m.GetSubstructMatch(core) for m in mt], useSVG=True )
img
```

```python
core_can = Chem.CanonSmiles(core_smi)
print(core_can)
```

```python
compare = 3
core_list = []
for i in range(0,compare-1):
    for j in range(i+1, compare):
        mt=[mols[i],mols[j]]
        mol_mcs = MCS.FindMCS(mt)
        #print(MCSResult)
        print(mol_mcs.numAtoms)
        print(mol_mcs.numBonds)
        #print(mol_mcs.smartsString)
        mcs_smart = mol_mcs.smartsString
        core = Chem.MolFromSmarts(mcs_smart)
        core_smi = Chem.MolToSmiles(core)
        print(core_smi)
        if core_smi not in core_list:
            core_list.append(core_smi)

print()
print(core_list)
```
