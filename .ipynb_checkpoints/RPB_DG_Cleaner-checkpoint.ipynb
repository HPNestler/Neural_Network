```python
import pandas as pd

```

```python
df = pd.read_excel('../Identified_interactions_sorted_by_process_DG.xlsx', sheet_name = None)

print(df)
```

```python
#print(len(df))
ldf = len(df)

dfpl = []

for i in df:
    print(i)

    if i !='All Combos':
        dfs = df[i]
        col = ['protein_1','protein_2']
        if i == 'stress_g_processing_b':
            col = ['protein1','protein2']
        #print(dfs[col])
        for j in range(0, len(dfs)):
            a = dfs.loc[j][0]
            b = dfs.loc[j][1]
            #print(a,b, type(a), type(b))
            if type(a) == str:
                #print(a)
                dfpl.append(a.upper())
            if type(b) == str:
                #print(b)
                dfpl.append(b.upper())
    #print()
    
print(dfpl)

dfp = pd.DataFrame(dfpl)
dfps = dfp.drop_duplicates()
print(dfps)

dfpl = dfps.values.tolist()
print(dfpl)
```
