```python
df_biomol = pd.concat([df1, df2], ignore_index=True)
print(df_biomol)
```

```python
df_hit = pd.read_excel('../PNAS_Paper/PNAS_hitlist.xlsx')

print(df_hit)
print()

hits = df_hit['name'].to_list()

print(hits)
```

```python
df_biohit = df_biomol[df_biomol['name'].isin(hits)]

print(df_biohit)
```

```python
hit_cas = []
hit_name = []

for i in range(0, len(df_biohit)):
    a = df_biohit.iloc[i]['CAS']
    b = df_biohit.iloc[i]['name']
    print(i, a, type(a), b)
    if type(a) == str:
        hit_cas.append(a)
    else:
        hit_name.append(b)

#print(df_hit)
ph = df_hit[df_hit['library'].isin(['Prestwick'])]
print(ph)
a = ph['name'].tolist()
print(a)
for i in a:
    hit_name.append(i)
        
print(hit_cas)
print(hit_name)
```

```python
dffl = 0

df_hits = pd.DataFrame(columns = ['name','smiles', 'cid'])

for i in hit_cas:
    result = pcp.get_compounds(i, 'name', as_dataframe = True)
    #x = len(result)
    x = 1
    for j in range(0, x):
        smi = result.iloc[j]['canonical_smiles']
        a = result.index[j]
        print(i, dffl, a, smi)#, result['canonical_smiles'])
        df_hits = df_hits.append({'name': i, 'smiles' : smi, 'cid' : a}, ignore_index = True)
    #print(a, smi)
    #df_hits = df_hits.append({'name': i, 'smiles' : smi, 'cid' : a}, ignore_index = True)
    if dffl == 0:
        df_struct = result
        dffl = 1
    else:
        df_struct = pd.concat([df_struct, result], ignore_index=True)
    
print()

for i in hit_name:
    try:
        result = pcp.get_compounds(i, 'name', as_dataframe = True)
        #print(i, dffl, result['canonical_smiles'])
        #x = len(result)
        x = 1
        for j in range(0, x):
            smi = result.iloc[j]['canonical_smiles']
            a = result.index[j]
            print(i, dffl, a, smi)#, result['canonical_smiles'])
            df_hits = df_hits.append({'name': i, 'smiles' : smi, 'cid' : a}, ignore_index = True)
        #print(a, smi)
        #df_hits = df_hits.append({'name': i, 'smiles' : smi, 'cid' : a}, ignore_index = True)
        #df_hits['name'] = i
        #df_hits['smiles'] = result['canonical_smiles']
        if dffl == 0:
            df_struct = result
            dffl = 1
        else:
            df_struct = pd.concat([df_struct, result], ignore_index=True)
    except:
        print()
        print(i, ' not found')
        print()
        
#print(df_struct)
print()
print(df_hits)
```
