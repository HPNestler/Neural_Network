```python
for i  in struct_ids:
    
    ### Remove Neural Network Entries
    sql = 'SELECT * FROM network.mol_net WHERE struct_id = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        print(len(a), a)
        sql = 'DELETE FROM network.mol_net WHERE struct_id = '+str(i)+';'
        print(sql)
        cur.execute(sql)
        db_chem.commit()
    print()
    
    ### Remove Fingerprints
    sql = 'SELECT * FROM public.structure_fingerprint WHERE structure_id = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        print(len(a), a)
        sql = 'DELETE FROM public.structure_fingerprint WHERE structure_id = '+str(i)+';'
        print(sql)
        cur.execute(sql)
        db_chem.commit()
    print()
    
    ### Remove Tanimotos
    sql = 'SELECT * FROM public.tanimoto WHERE source = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        print(len(a), a)
        sql = 'DELETE FROM public.tanimoto WHERE source = '+str(i)+';'
        print(sql)
        cur.execute(sql)
        db_chem.commit()
    print()
    
    sql = 'SELECT * FROM public.tanimoto WHERE target = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        print(len(a), a)
        sql = 'DELETE FROM public.tanimoto WHERE target = '+str(i)+';'
        print(sql)
        cur.execute(sql)
        db_chem.commit()
    print()
    
    ### Remove Calculated Properties
    sql = 'SELECT * FROM public.structure_prop_calc WHERE structure_id = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        print(len(a), a)
        sql = 'DELETE FROM public.structure_prop_calc WHERE structure_id = '+str(i)+';'
        print(sql)
        cur.execute(sql)
        db_chem.commit()
    print()
    
    ### Move and remove structure annotations
    sql = 'SELECT * FROM public.structure_properties sp WHERE sp.structure_id = '+str(i)+' and sp.property_type_id != 4 and sp.property_type_id != 5;'
    #print(sql)
    cur.execute(sql)
    a = cur.fetchall()
    if len(a) > 0:
        sql = 'SELECT smiles FROM public.structures sp WHERE sp.structure_id = '+str(i)+';'
        cur.execute(sql)
        smi = cur.fetchone()
        smi = str(smi[0])
        b = smi.split('.')
        c = ''
        for j in b:
            if len(j) > len(c):
                c = j
        print(len(a),a, smi)
        print(b, c)
        sql = 'select * from public.structures where molecule =\' '+c+' \'::mol;'
        print(sql)
        cur.execute(sql)
        d = cur.fetchall()
        print(d)
        for e in a:
            sql = 'UPDATE public.structure_properties SET structure_id = '+str(d[0][0])+' where property_id = '+str(e[3])+';'
            print(sql)
            cur.execute(sql)
            db_chem.commit()
    sql = 'DELETE FROM public.structure_properties WHERE structure_id = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    db_chem.commit()
    print()
        
    ### Finally Remove Structure
    sql = 'DELETE FROM public.structures WHERE structure_id = '+str(i)+';'
    print(sql)
    cur.execute(sql)
    db_chem.commit()
    print()    
    print('########### STRUCTURE REMOVED ########################')
    print()
        
        
```
