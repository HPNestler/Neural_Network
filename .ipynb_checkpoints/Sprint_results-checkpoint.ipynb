```python
import pandas as pd
import seaborn as sns; sns.set_theme(color_codes=True)
```

```python
df = pd.read_csv('../Sprint_Analysis_worker_1.csv')
print(df)
```

```python
result_list = []

for i in range(0, len(df)):
    exp_id = df.iloc[i]['exp_id']
    for j in range(8,16):
        res = df.iloc[i][j]
        print(i, exp_id, j, res, type(res))
        if res != 'na' and type(res) == str:
            a = [exp_id, j-7, res, 72, 'h', 'aurelien']
            print(a)
            result_list.append(a)
```

```python
columns = ['experiment_id', 'result_type', 'result_value', 'time_point', 'time_point_unit', 'assessor']

df_out = pd.DataFrame(result_list, columns=columns)

print(df_out)
```

```python
df_out.to_csv('../Sprint_Analysis_results_1.csv', index=False)
```
