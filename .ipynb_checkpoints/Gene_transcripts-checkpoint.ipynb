```python
blank = [0,1,8]
posc =  [22,23]
act_list = []
drop_list = []
na_list = []
stat_list = []
safe_ct = 0
tw = 56007

for i in range(4,tw):
    bl_l = []
    pc_l = []
    for j in blank:
        a = dfg.iloc[j,i]
        bl_l.append(a)
    for j in posc:
        a = dfg.iloc[j,i]
        pc_l.append(a)
        
    fl = dfg.iloc[:,i].to_list()
    #print(np.std(fl),np.mean(fl))
    
    if np.std(fl) <= np.mean(fl) and np.mean(fl) >10:
        #print(i, dfg.columns[i],' - ', np.std(bl_l),'/',np.mean(bl_l),np.std(pc_l),'/',np.mean(pc_l),np.std(fl),'/',np.mean(fl))
        act_list.append(i)
        b = [i, dfg.columns[i], np.std(bl_l),np.mean(bl_l),np.std(pc_l),np.mean(pc_l), np.std(fl), np.mean(fl)]
        #c = dfg.columns[i]
        stat_list.append(b)
    elif np.std(bl_l) < np.mean(bl_l) and np.mean(fl) > 10 and abs(np.mean(bl_l)-np.mean(fl))> (3*np.std(bl_l)):
        print(i, dfg.columns[i],' - ', np.std(bl_l),'/',np.mean(bl_l),np.std(pc_l),'/',np.mean(pc_l),np.std(fl),'/',np.mean(fl))
        act_list.append(i)
        b = [i, dfg.columns[i], np.std(bl_l),np.mean(bl_l),np.std(pc_l),np.mean(pc_l), np.std(fl), np.mean(fl)]
        #c = dfg.columns[i]
        stat_list.append(b)
        safe_ct = safe_ct+1
    else:    
        na_list.append(i)
        drop_list.append([i, dfg.columns[i], np.std(bl_l),np.mean(bl_l),np.std(pc_l),np.mean(pc_l), np.std(fl), np.mean(fl)])
        
print()
print(len(na_list), len(act_list),safe_ct)
#print(na_list)
```

```python
df2 = dfg.drop(dfg.iloc[:, na_list],axis = 1)
print(df2.shape)
print(df2.iloc[:,0:30])
```
