```python
import pandas as pd
from rdkit import Chem
#from rdkit.Chem import rdFMCS as MCS
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from datetime import datetime
import psycopg2
from rdkit.Chem.Draw import rdMolDraw2D
import IPython.display
```

```python
img_folder = '../mol_images'

print('Start Time: ',datetime.now())
print()

db_chem = psycopg2.connect(host = "10.2.0.140", dbname="chemistry", user="postgres", password="postgres")
#db_chem = psycopg2.connect(host = "192.168.86.34", dbname="chemistry", user="postgres", password="postgres")
cur = db_chem.cursor()
sql = "select zd.structure_id, ss.molecule from structure.zafrens_drugs zd join structure.structures ss on zd.structure_id = ss.structure_id where ss.svg_image is NULL order by structure_id;" #" and structure_id =27389;"
print(sql)
cur.execute(sql)
mollist = cur.fetchall()
print()
print('Images  to generate : ',len(mollist))
print()
print(mollist)
```

```python
#mols = []

for i in mollist:
    #print(i[0],type(i[1]))
    m = Chem.MolFromSmiles(i[1])
    #mols.append(m)
    #img_file = img_folder+"/zs_"+str(i[0])+".png"
    svg_file = img_folder+"/zs_"+str(i[0])+".svg"
    #Draw.MolToFile(m, img_file, size=(600,600))
    
    d = rdMolDraw2D.MolDraw2DSVG(600, 600)
    d.drawOptions().minFontSize = 16
    d.DrawMolecule(m)
    d.FinishDrawing()
    p = d.GetDrawingText()
    p = p.replace("rect style='opacity:1.0;fill:#FFFFFF","rect style='opacity:1.0;fill:none")

    #print(p)
  
    #i = IPython.display.SVG(p)
    #display(i)

    with open(svg_file, 'w') as f:
        f.write(p)
    
    sql = "UPDATE structure.structures SET svg_image = 1 where structure_id ="+str(i[0])+";"
    cur.execute(sql)
    
#print(mols)

#img = Draw.MolsToGridImage(mols, molsPerRow=5, maxMols = 100, useSVG=True )
#img.show()

db_chem.commit()
db_chem.close()

print()
print('End Time: ',datetime.now())
```
