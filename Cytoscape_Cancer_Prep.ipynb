```python
pubcid = raw_df['PubCHEM']
pubcid = pubcid.drop_duplicates()
#print(pubcid)
m = pubcid.to_list()
#print(m)
publist = []
for i in m:
    #print(i, type(i))
    if type(i) == str:
        k = i.split(',')
        for l in k:
            try:
                j = int(l)
                #print(j, type(j))
                publist.append(j)
            except:
                print('No Convert : ', l)

#print(publist)
complist = []
smilist = []

for i in publist:
    a = pcp.Compound.from_cid(i)
    print(i)
    b = a.canonical_smiles
    print(a, b)
    complist.append(a)
    smilist.append(b)
```

```python
#print(smilist)
for i in smilist:
    j = i.split('.')
    print(i,j)
    for k in j:
        m = (k,)
        print(k, m)
        cp.SmilesInsert(m)
```

```python
#print(raw_df)

n = []

for i in range(0, len(raw_df)):
    j = raw_df["PubCHEM"].iloc[i] 
    #print(j)
    if type(j) == float or j == 'none' or j == 'several':
        k = raw_df["Name"].iloc[i]
        #print('Name: ',k)
        if k not in n:
            n.append(k)
        k = raw_df["Synonyms"].iloc[i] 
        if type(k) == str:
            #print('Syn : ',k)
            k = k.split(',')
            for l in k:
                #print(l)
                if l not in n:
                    n.append(l)
 
#print(n)
c =[]
namecomp = []
namesmi = []

for i in n:
    a = pcp.get_compounds(i, 'name')
    print(i, a)
    for j in a:
        cx = j.cid
        print(cx)
        if cx not in c:
            c.append(cx)
            namecomp.append(j)
            s = j.canonical_smiles
            print(s)
            namesmi.append(s)
            
print(namecomp)
print(namesmi)
```

```python
print(namesmi)
for i in namesmi:
    j = i.split('.')
    print(i,j)
    for k in j:
        m = (k,)
        print(k, m)
        cp.SmilesInsert(m)
```

```python


#print(raw_df['Name'])

nod_list = []
#nod_tar = []
#nod_path = []
edge_list = []
edge_ct = []
edge_tp = []

for i in range(0,len(raw_df)):
    d = raw_df['Source'].iloc[i]
    d = 'DX_'+ str(d)
    l = raw_df['Name'].iloc[i]
    ci = raw_df['PubCHEM'].iloc[i]
    si = raw_df['SMILES'].iloc[i]
    t = raw_df['Targets'].iloc[i] 
    p = raw_df['Target pathway'].iloc[i]
    p = p.replace(',','/')
    print(d, l ,t, p)
    nc = [d, l, 'Compound', ci, si]
    if nc not in nod_list:
        nod_list.append(nc)
    np = [p, p, 'Pathway']
    if np not in nod_list:
        nod_list.append(np)
    print(nc)
    print(np)
    ts = t.split(',')
    for j in ts:
        j = j.strip()
        nt = (j, j, 'Target')
        if nt not in nod_list:
            nod_list.append(nt)
        print(nt)
        ect = [d, 'ct', j, 1]
        if ect not in edge_list:
            edge_list.append(ect)
        print(ect)
        etp = [j, 'tp', p, 1]
        if etp not in edge_list:
            edge_list.append(etp)
        print(etp)
    print()
    
print(nod_list)
print()
print(edge_list)
```

```python
nodes = pd.DataFrame(nod_list, columns = ['Name', 'Label', 'Type', 'PubCID', 'SMILES'])
edges = pd.DataFrame(edge_list, columns = ['Source','Interaction','Target','Weight'])

print(nodes)
print()
print(edges)

nodes.to_csv('../Cancer_Nodes.csv', index = False)
#edges.to_csv('../Cancer_Edges.csv', index = False)
```
