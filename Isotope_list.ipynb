```python
#print(isomass.nist_mass)

nm = isomass.crc_mass
mz_prop = 0.0001
#print(type(nm))
#print(nm)

isolist = {}
for e,l in nm.items():
    #print(e,l)
    isoel = {}
    for n,m in l.items():
        if n !=0 and m[1]!=0:
            #print(e,n,m)
            if m[1] >= mz_prop:
                isoel[n]=m
    if len(isoel) !=0:
        isolist[e] = isoel
        
print(isolist)

```

```python
mol =['Cl','Cl','Cl', 'C','H','C','H','H']

mz_list = [(0.0,1.0)]
print('Starting Enumeration:',mz_list)

for i in mol:
    nl = []
    atom = isolist[i]
    print('Atom:',i,atom)
    for j in mz_list:
        for k,v in atom.items():
            #print(j,k,v)
            ms = round(j[0]+v[0],6)
            ps = j[1]*v[1]
            #print(ms,ps)
            a = (ms,ps)
            nl.append(a)
    print('Raw Enumeration:',nl)
   
    ### Purging Double MZ ###
    nln = len(nl)-1
    #print(nln)
    i = 0
    while i < nln:
        j = i+1
        while j <= nln:
            #print(i, j, nl[i], nl[j])
            if nl[i][0] == nl[j][0]:
                a = nl[i][1] + nl[j][1]
                b = (nl[i][0], a)
                nl[i] = b
                c = nl.pop(j)
                nln = len(nl)-1
                j = j-1
                #print(a,b,c,nln)
                print(i, j, nl[i], nl[j])
                print('Double Enumeration:',nl)
            j = j+1
        i=i+1
    ### Purging Low Level MZ ###
    nln = len(nl)-1
    i = 0
    while i <= nln:
        a = nl[i][1]
        if a < mz_prop:
            c = nl.pop(i)
            print('Propability Drop:',c)
            nln = len(nl)-1
            i = i-1
        i = i+1
    
    print()
    print('Starting Enumeration:',nl)
    mz_list = nl
    
#print(mz_list)
    
```
