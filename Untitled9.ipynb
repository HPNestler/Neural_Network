```python
import xmltodict as xtd
import pprint
import pandas as pd
```

```python
##Wuxi_Targets_220510


  
# Open the file and read the contents
with open('../Wuxi_Targets_220510.mrv', 'r', encoding='utf-8') as file:
    my_xml = file.read()
  
# Use xmltodict to parse and convert 
# the XML document
my_dict = xtd.parse(my_xml)
  
# Print the dictionary
pprint.pprint(my_dict, indent=1)
```

```python
mol = my_dict['cml']['MDocument']['MChemicalStruct']['molecule']
#print(mol)
#pprint.pprint(mol, indent=1)
atoms = mol['atomArray']['atom']
pprint.pprint(atoms, indent=1)
```

```python
atom_list = []

for i in atoms:
    print(i)
    a = []
    for k,v in i.items():
        #print(k,v)
        if k in ['@id','@elementType','@x2', '@y2']:            
            a.append(v)
    print(a)
    atom_list.append(a)
    
atom_df = pd.DataFrame(atom_list, columns = ['ID', 'Element', 'X','Y'])
```

```python
print(atom_df)
```

```python
bonds = mol['bondArray']['bond']
#pprint.pprint(bonds, indent=1)

bond_list = []

for i in bonds:
    print(i)
    a = []
    for k,v in i.items():
        #print(k,v)
        if k in ['@id','@order']:
            a.append(v)
        elif k == '@atomRefs2':
            j = v.split(' ')
            #print(j)
            for l in j:
                a.append(l)
    print(a)
    bond_list.append(a)
    
bond_df = pd.DataFrame(bond_list, columns = ['ID', 'A1', 'A2','Type'])
```

```python
print(bond_df)
```
