```python
import numpy as np
import math
import psycopg2

# open database connection
db_nn = psycopg2.connect(host = "localhost", dbname="Chemistry", user="postgres", password="postgres")

cur = db_nn.cursor()
cur.execute('SELECT "Fingerprint_ID", "FP_Shannon" FROM public."Fingerprints" WHERE "FP_Type" = 1 ORDER BY "Fingerprint_ID"')
fp_list = cur.fetchall()
print(fp_list)

fp = len(fp_list)
print('Fingerprints : ',fp)
#fp_list.sort()


ms = 3
# fp = 4

net = np.zeros((ms, ms, fp, 2))

# generate random floating point values
from random import seed
from random import random
# seed random number generator
seed()

r = 500
for i in range(0,r):
    x = int(ms*random())
    y = int(ms*random())
    f = int(fp*random())
    # t = int(2*random())
    v = int(10*random())
    # print(x,y,f,t,v)
    net[x][y][f][0] = v

print(net)

fit = np.zeros((ms,ms))
# a = math.sqrt(4)
print()
print(fit)

# probe = [1,3,5,7]
probe =[]
for i in range(0,fp):
    v = int(10*random())
    probe.append(v)
    
print()
print('Probe : ', probe)

pi = math.pi

for i in range(0, ms):
    for j in range(0,ms):
        v = 0
        for k in range(0, fp):
            # print(i,j,k)
            p = probe[k]
            n = net[i][j][k][0]
            # s = net[i][j][k][1]+.5
            s = fp_list[k][1]
            # d = math.sqrt((p-n)**2)
            d = 1/(2*pi*pow(s,2)) * math.exp(-(pow(p-n,2))/(2*pow(s,2)))
            v = v + d
             
        fit[i][j] = v
        
print()
print(fit)

m = np.max(fit)
print(m)
ind = np.unravel_index(np.argmax(fit, axis=None), fit.shape)
print(ind)

```
