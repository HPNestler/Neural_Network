```python
import requests, sys
 
server = "https://rest.ensembl.org"
ext = "/xrefs/id/ENSG00000157764?external_db=Uniprot_gn"
 
r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
if not r.ok:
  r.raise_for_status()
  sys.exit()
 
decoded = r.json()
print(repr(decoded))

print(decoded)
```
