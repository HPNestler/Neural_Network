```python
import math

def dist(i,j):
    #print(i, j)
    li = len(i)
    lj = len(j)
    q = 0
    if li != lj:
        print('Unequal Length')
    else:
       #q = 0
        for k in range(0,li):
            a = math.sqrt((i[k]-j[k])**2)
            q = q + a
            # print(q, a) 
        return q

ic = 0
for i in kmeans.cluster_centers_:
    #print(i)
    jc = 0
    for j in kmeans.cluster_centers_:
        #print(i, j)
        q = dist(i,j)
        print(ic,jc, q)
        jc = jc + 1
    ic = ic + 1
```
