```python
import pandas as pd
```

```python
df = pd.read_excel('../Chemicals_inventory.xlsx')
print(df)
```

```python
inv_list = []
bc = 1

for i in range(0, len(df)):
    a = df.iloc[i][0]
    b = df.iloc[i][1]
    c = df.iloc[i][3]
    if type(c) != float:
        if type(c) == str:
            d = df.iloc[i][2]+' / '+df.iloc[i][3]
        else:
            d = df.iloc[i][2]+' / '+str(df.iloc[i][3])
    else:
        d = df.iloc[i][2]
        
    e =  df.iloc[i][4]
    ef = e.find(' x ')
    em = 1
    eb = e
    if ef > 0:
        em = int(e[0:ef])
        eb = e[ef+3:]
        #print(e, em, eb)
    eu = ''
    ea = ''
    for j in eb:
        #print(j, ord(j))
        if ord(j)>47 and ord(j)<58:
            ea = ea + j
        if ord(j)>64:
            eu = eu + j
    eu = eu.lower()
    #print(a,b,d, type(c))
    #print(a,b,d, ea, eu, em)
    et ='bottle'
    if eu == 'plate':
        et = 'plate'
        ea = ''
        eu = ''
    
    for j in range(0, em):
        qr = 'A'+str(bc).zfill(7)
        q = [a, b, d, qr, et, ea, eu] 
        print(q)
        inv_list.append(q)
        bc = bc + 1
        
#print(inv_list)
```

```python
headers = ['CONTAINER ID', 'CONTENT', 'COMMENTS', 'ALTERNATE', 'TYPE ID', 'AMOUNT', 'UNIT']

inv_df = pd.DataFrame(inv_list, columns=headers)
           
print(inv_df)
```

```python
inv_df.to_csv('../inventory.csv')
```
